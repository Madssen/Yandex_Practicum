Рейтинг платформ по суммарным продажам:

```python
data_sales = data.groupby(['platform']).agg({'sum_sales':'sum'}).sort_values('sum_sales', ascending=False)
data_sales.head(10)
```

Возьму 5 топовых платформ для дальнейшего анализа.

```python
data_sales_5 = (
    data.query('platform in @top_platforms')
    .groupby(['platform', 'year_of_release']).agg({'sum_sales':'sum'})
    .sort_values(by='sum_sales', ascending=False)
)
data_sales_5.head()
```

```python
data_sales_5_pivot = pd.pivot_table(data_sales_5, index = 'year_of_release', columns='platform', values='sum_sales')
data_sales_5_pivot.head()
```

## Шаг 5. Проверьте гипотезы

### Cредние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Нулевая гипотеза: рейтинги платформ Xbox One и PC равны


Альтернативная гипотеза: рейтинги платформ Xbox One и PC не равны

```python
data_mean = (
    data_2013.query('user_score >= 0 & (platform == "XOne" or platform == "PC")')
    .groupby('platform')['user_score'].agg('mean')
)
data_mean
```

```python
data_Xone = data_2013.query('platform == "XOne" & user_score >=0')['user_score']
data_PC = data_2013.query('platform == "PC" & user_score >=0')['user_score']
```

```python
sample_1 = data_Xone
sample_2 = data_PC

alpha = .1 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») равны


Альтернативная гипотеза: Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») не равны

```python
data_mean = (
    data_2013.query('user_score >= 0 & (genre == "Action" or genre == "Sports")')
    .groupby('genre')['user_score'].agg('mean')
)
data_mean
```

```python
data_Action = data_2013.query('user_score >= 0 & genre == "Action"')['user_score']
data_Sports = data_2013.query('user_score >= 0 & genre == "Sports"')['user_score']
```

```python
sample_1 = data_Action
sample_2 = data_Sports

alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    sample_1, 
    sample_2)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```
